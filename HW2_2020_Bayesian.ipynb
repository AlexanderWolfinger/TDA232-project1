$$


Using the definition of the l2 norm with the last step, we get

$$
\sum_{i=1}^n (y_i-\mathbf{w}^\top \mathbf{x}_i)^2 = \left( \mathbf{y} - \mathbf{X} \mathbf{w} \right)^\top \left( \mathbf{y} - \mathbf{X} \mathbf{w} \right) = |\mathbf{y} - \mathbf{X} \mathbf{w}|_2^2
$$

and hence, the full expression is given as

$$
\mathcal{L}(\mathbf{w}) = \left(\sum_{i=1}^n (y_i - \mathbf{w}^T\mathbf{x}_i)^2 \right) + \left(\alpha \sum_{j=1}^p w_j^2 \right) = |\mathbf{y} - \mathbf{X} \mathbf{w}|_2^2 + \alpha |\mathbf{w}|_2^2.
$$

2. Deriving $\frac{d}{d\mathbf{w}}\mathcal{L}(\mathbf{w})$:
$$
\begin{align}
\frac{d}{d\mathbf{w}}\mathcal{L}(\mathbf{w}) &= \frac{d}{d\mathbf{w}} \left( |\mathbf{y} - \mathbf{X}\mathbf{w}|_2^2 + \alpha|\mathbf{w}|_2^2 \right) \\
&= \frac{d}{d\mathbf{w}} \left( \left(\mathbf{y} - \mathbf{X}\mathbf{w}\right)^\top \left(\mathbf{y} - \mathbf{X}\mathbf{w}\right) + \alpha \mathbf{w}^\top \mathbf{w} \right) \\
&= \frac{d}{d\mathbf{w}} \left( \left(\mathbf{y}^\top - (\mathbf{X}\mathbf{w})^\top\right) \left(\mathbf{y} - \mathbf{X}\mathbf{w}\right) + \alpha \mathbf{w}^\top \mathbf{w} \right) \\
&= \frac{d}{d\mathbf{w}} \left( \left(\mathbf{y}^\top - \mathbf{w}^\top \mathbf{X}^\top \right) \left(\mathbf{y} - \mathbf{X}\mathbf{w}\right) + \alpha \mathbf{w}^\top \mathbf{w} \right) \\
&= \frac{d}{d\mathbf{w}} \left( \mathbf{y}^\top \mathbf{y} - \mathbf{y}^\top \mathbf{X} \mathbf{w} - \mathbf{w}^\top \mathbf{X}^\top \mathbf{y} + \mathbf{w}^\top \mathbf{X}^\top \mathbf{X} \mathbf{w} + \alpha \mathbf{w}^\top \mathbf{w} \right) \\
&= 0 - (\mathbf{y}^\top \mathbf{X})^\top - \mathbf{X}^\top \mathbf{y} + 2 \mathbf{X}^\top \mathbf{X} \mathbf{w} + 2 \alpha \mathbf{w} \\
&= - \mathbf{X}^\top \mathbf{y} - \mathbf{X}^\top \mathbf{y} + 2 \mathbf{X}^\top \mathbf{X} \mathbf{w} + 2 \alpha \mathbf{w} \\
&= 2 \mathbf{X}^\top \mathbf{X} \mathbf{w} - 2 \mathbf{X}^\top \mathbf{y} + 2 \alpha \mathbf{w}.
\end{align}
$$


3. Since $\mathcal{L}(\mathbf{w})$ is convex in $\mathbf{w}$, there exists a min-point $\mathbf{w}^{*}$ s.t. $\mathcal{L}(\mathbf{w}^{*}) = \mathcal{L}^{*}$. So $\frac{\partial \mathcal{L(\mathbf{w})}}{\partial \mathbf{w}} = 0$ will be satisfied for $\mathbf{w} = \mathbf{w}^{*}$.


\begin{align*}
\frac{\partial \mathcal{L(\mathbf{w})}}{\partial \mathbf{w}} = 2 \mathbf{X}^\top \mathbf{X} \mathbf{w} - 2 \mathbf{X}^\top \mathbf{y} + 2 \alpha \mathbf{w} &= 0 \\
\Leftrightarrow \mathbf{X}^\top \mathbf{X} \mathbf{w} + \alpha \mathbf{w} &= \mathbf{X}^\top \mathbf{y} \\
\Leftrightarrow ( \mathbf{X}^\top \mathbf{X} + \alpha \mathbf{I} ) \mathbf{w} &= \mathbf{X}^\top \mathbf{y} \\
\Rightarrow \mathbf{w}^{*} = ( \mathbf{X}^\top \mathbf{X} + \alpha \mathbf{I} )^{-1} \mathbf{X}^\top \mathbf{y}
\end{align*}
where $\mathbf{I}$ is the unit matrix of dimensions $p \times p.$

4. As mentioned, $\mathcal{L}(\mathbf{w})$ is a convex function meaning that if $\mathbf{w}^*$ is well defined, it is a unique global minimum. From 3. we see that computation of $\mathbf{w}^*$ requires the inverse of $\mathbf{M} \in \mathbb{R}^{p \times p} = ( \mathbf{X}^\top \mathbf{X} + \alpha \mathbf{I} )^{-1}$. Since any positive definite matrix is invertible (specifically, any square matrix with only non-zero eigenvalues is invertible), we can choose the value $\alpha$ to ensure that this is always the case. Solving for eigenvalues

$$
    \text{det}(\mathbf{M})=\text{det}(\mathbf{X}^\top \mathbf{X}-\gamma \mathbf{I})=0
$$

where $\gamma=(\lambda-\alpha)$. This yields the eigenvalues $\gamma_1, \dots, \gamma_p$. For the solution to be strictly positive for all $\gamma_i$, we require that $\text{min}(\gamma_i) > 0 \Leftrightarrow \text{min}(\lambda_i - \alpha) > 0$ and therefore choose $\alpha$ to satisfy $\text{min}(\lambda_i) > \alpha$ where $\lambda_i$, $i = 1, \dots, p$, are the solutions to the equation 
$$
    \text{det}(\mathbf{X}^\top \mathbf{X}-\lambda \mathbf{I})=0.
$$

Hence the matrix $\mathbf{M}$ is full rank and positive definite and the matrix inverse $\mathbf{M}^{-1}$ exists. This implies that the solution $\mathbf{w}^{*}$ exists and is a unique global minimum.

```python
# Make it possible to show plots in the notebooks.
%matplotlib inline
import numpy as np
from sklearn import datasets
import matplotlib
import matplotlib.pyplot as plt


def fit_linear_with_regularization(X, y, alpha):
    """Estimates w* by using w* = (X^T * X + alpha * I)^-1 * X^T * y"""
    p = np.size(X,1) # size p of matrix X (n x p)
    X_transpose_X = np.matmul(np.transpose(X),X) # calculates: X^T * X
    alpha_identity = alpha*np.identity(p) # calculates: alpha * I (I is the unit matrix of size p)
    inverse = np.linalg.inv(X_transpose_X + alpha_identity) # calculates the invers of the sum of the above two
    X_transpose_y = np.matmul(np.transpose(X),y) # calculates: X^T * y
  
    w_star = np.matmul(inverse, X_transpose_y) # calculates: w* by multiplying the inverse and X^T * y
    return w_star # returns the estimation of w

def predict(X_test, w):
    """Predict function that calculates y = X_test * w"""
    y_pre = np.matmul(X_test,w) # calculates y = X_test * w
    return y_pre # returns the prediction of y


def plot_prediction(X_test, y_test, y_pred):
    """Plot data set and prediction, returns MSE"""
    # Scatter plot the first feature of X_test (x-axis) and y_test (y-axis)
    ax[1].plot(X_test[:,1],y_test,'r.', label="datapoints")
  
    # Plot y_pred using the first feature of X_test as x-axis
    ax[1].plot(X_test[:,1],y_pred,'b*', label="predictions") # not linear since y_pred is dependent on X[:,:]
    
    # add legend to plot
    ax[1].legend()
    ax[1].set(xlabel = 'X[1]', ylabel = 'y')
    ax[1].set_title('Data set and prediction')
  
    # Compute the mean squared error as MSE = (1/n) sum over i (y_i - y_pred_i)^2
    mean_squared_error = np.mean((y_test - y_pred)**2)
    
    return mean_squared_error # returns MSE


def find_alpha(X_test, X_train, y_train, ax, a_min, a_max):
    """Finding the alpha that minimizes MSE, plots MSE for the alpha vector"""
    l = 10000 # number of alphas to test
    alphas = np.linspace(a_min, a_max, l) # interval for alpha
    errors = np.zeros(l)
    for i in range(l): # Assuming for-loop is ok for the bonus question
        alpha = alphas[i] # for each alpha value in interval

        # Train using linear regression with regularization and find optimal model
        w = fit_linear_with_regularization(X_train, y_train, alpha)

        # Plots and mean squared error
        MSE = np.mean((y_test - np.matmul(X_test,w))**2)
        errors[i] = MSE
    
    ax[0].plot(alphas,errors)
    ax[0].set(xlabel = '$\\alpha$', ylabel = 'MSE')
    ax[0].set_title('Finding optimal $\\alpha$')
    return alphas[np.argmin(errors)] # return optimal alpha


# Load the diabetes dataset
X, y = datasets.load_diabetes(return_X_y=True)

# Normalize the dataset
# [0,1] min-max normalized data
# Compute min and max for features
Xmin = np.min(X,axis=0)
Xmax = np.max(X,axis=0)
# Compute normalization factor and repeat into matrix
nf = (1/(Xmax-Xmin))
nf = np.repeat(nf[np.newaxis,:],X.shape[0],axis=0)
# Multiply each element with normalization factor
X = (X-np.repeat(Xmin[np.newaxis,:],X.shape[0],axis=0))*nf


X = np.c_[ np.ones((np.size(X,0),1)), X ] # Adding a column of ones to X

# Define figures
fig,ax = plt.subplots(1,2,figsize = (15,5))


# Split the dataset into training and test set
num_test_elements = 20

X_train = X[:-num_test_elements]
X_test = X[-num_test_elements:]

y_train = y[:-num_test_elements]
y_test = y[-num_test_elements:]


# Finding best alpha
#a_min = 1e-4
#a_max = 0.02
a_min = 0
a_max = 0.25
alpha = find_alpha(X_test, X_train, y_train, ax, a_min, a_max) # using optimal alpha
print('Optimal alpha for given data set and normalized X is ', alpha)


# Train using linear regression with regularization and find optimal model
w = fit_linear_with_regularization(X_train, y_train, alpha)


# Make predictions using the testing set X_test
y_pred = predict(X_test, w)


# Plots and mean squared error
error = plot_prediction(X_test, y_test, y_pred)
print('Mean Squared error is ', error)

# Show the plot
plt.show();
```

By Bayes rule we have that the posterior $$p(\mathbf{w|y,X}) = \frac{p(\mathbf{y|X,w})p(\mathbf{w})}{p(\mathbf{y|X})}$$ where $p(\mathbf{y|X})$ is a normalisation constant that can be ignored for our purposes.

The $p$-dimensional multivariate normal distribution $\mathcal{N}(\mathbf{\mu}, \mathbf{\Sigma})$ is given by
$$
p(\mathbf{x}) = \frac{1}{(2\pi)^{p/2}|\mathbf{\Sigma}|^{1/2}} \exp\left( -\frac{1}{2}(\mathbf{x - \mu})^T \mathbf{\Sigma}^{-1}(\mathbf{x - \mu}) \right)
$$
for some mean $\mathbf{\mu}$ and covariance matrix $\mathbf{\Sigma}$.

For the prior distribution we have
$$
p(\mathbf{w}) = \mathcal{N}(\mathbf{0}, \lambda^{-1} \mathbf{I}) = \frac{1}{(2\pi)^{p/2}|\lambda^{-1}\mathbf{I}|^{1/2}} \exp \left( -\frac{1}{2} (\mathbf{w - 0})^T (\lambda^{-1}\mathbf{I})^{-1}(\mathbf{w - 0}) \right) = \{ |c\mathbf{I}| = c|\mathbf{I}| = c\cdot 1 \} = \frac{\lambda^{1/2}}{(2\pi)^{p/2}1^{1/2}} \exp \left( -\frac{\lambda}{2} \mathbf{w}^T \mathbf{I}^{-1}\mathbf{w} \right) = \{ \mathbf{I}^{-1} = \mathbf{I}, \; \mathbf{Ix} = \mathbf{x} \} = \sqrt{\frac{\lambda}{(2\pi)^p}} \exp \left( -\frac{\lambda}{2} \mathbf{w}^T \mathbf{w} \right).
$$

For the univariate normally distributed case we have that $\Sigma \rightarrow \sigma^2$ and $p=1$. This results in the following expression for the likelihood.

$$
p(\mathbf{y} | \mathbf{X}, \mathbf{w}) = \prod_{i=1}^n \mathcal{N}(\mathbf{w}^T \mathbf{x}_i, \gamma^{-1}) = \prod_{i=1}^n \frac{1}{(2\pi)^{1/2}\gamma^{-1/2}} \exp \left( -\frac{1}{2} \frac{(y_i - \mathbf{w^Tx}_i)^2}{\gamma^{-1}} \right) = \prod_{i=1}^n \sqrt{\frac{\gamma}{2\pi}} \exp \left( -\frac{\gamma}{2} (y_i - \mathbf{w^Tx}_i)^2 \right)
$$

Ignoring the normalisation constant in the denominator of bayes rule we get for the posterior 
$$
p(\mathbf{w|y,X}) \propto p(\mathbf{y|X,w})p(\mathbf{w}) = \prod_{i=1}^n \sqrt{\frac{\gamma}{2\pi}} \exp \left( -\frac{\gamma}{2} (y_i - \mathbf{w^Tx}_i)^2 \right) \sqrt{\frac{\lambda}{(2\pi)^p}} \exp \left( -\frac{\lambda}{2} \mathbf{w}^T \mathbf{w} \right) = \sqrt{\frac{\gamma \lambda}{(2 \pi)^{p+1}}} \exp \left( -\frac{\lambda}{2} \mathbf{w}^T \mathbf{w} \right) \prod_{i=1}^n \exp \left( -\frac{\gamma}{2} (y_i - \mathbf{w^Tx}_i)^2 \right).

$$
\ln{p(\mathbf{w}|\mathbf{y},\mathbf{X})} \propto -\frac{\lambda}{2}|\mathbf{w}|_2^2-\frac{\gamma}{2}|\mathbf{y}-\mathbf{X} \mathbf{w}|_2^2 \\
-\frac{2}{\gamma}\ln{p(\mathbf{w}|\mathbf{y},\mathbf{X})} \propto  \frac{\lambda}{\gamma}|\mathbf{w}|_2^2+|\mathbf{y}-\mathbf{X} \mathbf{w}|_2^2.
$$
If we choose $\alpha=\lambda/\gamma$, then
$$
-\frac{2}{\gamma}\ln{p(\mathbf{w}|\mathbf{y},\mathbf{X})} \propto  \alpha|\mathbf{w}|_2^2+|\mathbf{y}-\mathbf{X} \mathbf{w}|_2^2 = \mathcal{L}(\mathbf{w}) \\
\ln{p(\mathbf{w}|\mathbf{y},\mathbf{X})} \propto -\frac{\gamma}{2}\mathcal{L}(\mathbf{w}).
$$
Hence, minimizing the function $\mathcal{L}(\mathbf{w})$ maximizes the posterior due to minus sign that turns a convex function into a concave and vice versa when multiplied by it.
